```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
algo = ['1 thread', '1+ thread', 'Builtin', 'Builtin parallel']
rust = [20.12, 10.34, 8.82, 4.25]
cpp  = [16.16, 9.86, 9.00, 3.21]
go   = [20.69, 10.00, 30.29, 0]
xpos = np.arange(4)*2

plt.bar(xpos-0.4, rust, width=0.4, label="Rust")
plt.bar(xpos,     cpp,  width=0.4, label="C++")
plt.bar(xpos+0.4, go,   width=0.4, label="Go")

plt.xticks(xpos, algo)
plt.ylabel("Time (s)")
plt.title('Sorting speed in compiled languages')
plt.legend()
```
